```python
import numpy as np
import matplotlib.pylab as plt
```

```python
tau_timeDiff = np.loadtxt('20200728_timeDifference_tau.csv')
e_timeDiff = np.loadtxt('20200728_timeDifference_e.csv')
e_pVal = np.loadtxt('20200728_pval_e.csv')
tau_pVal = np.loadtxt('20200728_pval_tau.csv')
```

```python
plt.plot(abs(tau_timeDiff), abs(np.log(tau_pVal)), '.', label='tau')
plt.plot(abs(e_timeDiff), abs(np.log(e_pVal)), '.', label = 'e')
plt.legend()
plt.xlabel('Time Difference (ns)')
plt.ylabel('p-val(double peak)/p-val(bifurcated gaussian)')
plt.savefig('timediffVSpval.pdf', dpi = 200)
```

```python
tau_timeDiff = np.loadtxt('20200728_timeDifference_tau.csv')
e_timeDiff = np.loadtxt('20200728_timeDifference_e.csv')
e_LRR = np.loadtxt('20200728_LRR_e.csv')
tau_LRR = np.loadtxt('20200728_LRR_tau.csv')

plt.plot(abs(tau_timeDiff), abs(np.log(tau_LRR)), '.', label = 'tau')
plt.plot(abs(e_timeDiff), abs(np.log(e_LRR)), '.', label = 'e')
plt.legend()
plt.xlabel('Time Difference (ns)')
plt.ylabel('LikelihoodR(double peak)/LikelihoodR(bifurcated gaussian)')
plt.savefig('timediffVSLRR.pdf', dpi = 200)
```

```python
tau_timeDiff = np.loadtxt('20200728_timeDifference_tau.csv')
e_timeDiff = np.loadtxt('20200728_timeDifference_e.csv')
e_wid1_wid2 = np.loadtxt('20200728_wid1_wid2_e.csv')
tau_wid1_wid2 = np.loadtxt('20200728_wid1_wid2_tau.csv')

plt.plot(abs(tau_timeDiff), abs(np.log(tau_wid1_wid2)), '.', label = 'tau')
plt.plot(abs(e_timeDiff), abs(np.log(e_wid1_wid2)), '.', label = 'e')
plt.legend()
plt.xlabel('Time Difference (ns)')
plt.ylabel('wid1/wid2')
plt.savefig('timediffVSwid.pdf', dpi = 200)
```

```python
tau_timeDiff = np.loadtxt('20200728_timeDifference_tau.csv')
e_timeDiff = np.loadtxt('20200728_timeDifference_e.csv')
e_amp1_amp2 = np.loadtxt('20200728_amp1_amp2_e.csv')
tau_amp1_amp2 = np.loadtxt('20200728_amp1_amp2_tau.csv')

plt.plot(abs(tau_timeDiff), abs(np.log(tau_amp1_amp2)), '.', label = 'tau')
plt.plot(abs(e_timeDiff), abs(np.log(e_amp1_amp2)), '.', label = 'e')
plt.legend()
plt.xlabel('Time Difference (ns)')
plt.ylabel('wid1/wid2')
plt.savefig('timediffVSamp.pdf', dpi = 200)
```

```python
bins_tau = np.arange(min(abs(tau_timeDiff)), max(abs(tau_timeDiff))+1, 1)
bins_e = np.arange(min(abs(e_timeDiff)), max(abs(e_timeDiff))+1, 1)
plt.hist(abs(tau_timeDiff), bins = bins_tau, histtype='step', log=True, normed=True, label='tau')
plt.hist(abs(e_timeDiff), bins = bins_e, histtype='step', log=True, normed=True, label='e')
plt.legend()
plt.xlabel('Double peak amplitude ratio')
plt.savefig('timediffHist.pdf', dpi=200)
```
