```python
read_file = open('/data/p-one/akatil/analysis/RecoPulses/RecoPulseFitInfo.csv', 'r')

file_num, frame_num, lepton, DOM, string = ([]), ([]), ([]), ([]), ([])
binEntries_mean, success_biGauss, success_doublePeak = ([]), ([]), ([])
biGauss_pos, biGauss_wid, biGauss_rat, biGauss_amp = ([]), ([]), ([]), ([])
doublePeak_amp1, doublePeak_amp2, doublePeak_pos1, doublePeak_pos2 = ([]), ([]), ([]), ([])
doublePeak_rat1, doublePeak_rat2, doublePeak_wid1, doublePeak_wid2 = ([]), ([]), ([]), ([])
area_data, area_biGauss_fit, area_doublePeak_fit = ([]), ([]), ([])
gof_biGauss, gof_doublePeak = ([]), ([])
with read_file:

    reader = csv.DictReader(read_file)
    
    for row in reader:
        file_num = np.append(file_num, row['file'])
        frame_num = np.append(frame_num, row['frame'])
        lepton =np.append(lepton, row['lepton_type'])
        DOM = np.append(DOM, row['DOM'])
        string = np.append(string, row['string'])
        binEntries_mean = np.append(binEntries_mean, float(row['binEntries_mean']))
        success_biGauss = np.append(success_biGauss, row['success_biGauss'])
        success_doublePeak = np.append(success_doublePeak, row['success_doublePeak'])
        biGauss_pos = np.append(biGauss_pos, float(row['biGauss_pos']))
        biGauss_wid = np.append(biGauss_wid, float(row['biGauss_wid']))
        biGauss_rat = np.append(biGauss_rat, float(row['biGauss_rat']))
        biGauss_amp = np.append(biGauss_amp, float(row['biGauss_amp']))
        doublePeak_pos1 = np.append(doublePeak_pos1, float(row['doublePeak_pos1']))
        doublePeak_wid1 = np.append(doublePeak_wid1, float(row['doublePeak_wid1']))
        doublePeak_rat1 = np.append(doublePeak_rat1, float(row['doublePeak_rat1']))
        doublePeak_amp1 = np.append(doublePeak_amp1, float(row['doublePeak_amp1']))
        doublePeak_pos2 = np.append(doublePeak_pos2, float(row['doublePeak_pos2']))
        doublePeak_wid2 = np.append(doublePeak_wid2, float(row['doublePeak_wid2']))
        doublePeak_rat2 = np.append(doublePeak_rat2, float(row['doublePeak_rat2']))
        doublePeak_amp2 = np.append(doublePeak_amp2, float(row['doublePeak_amp2']))
        area_data = np.append(area_data, float(row['area_data']))
        area_biGauss_fit = np.append(area_biGauss_fit, float(row['area_biGauss_fit']))
        area_doublePeak_fit = np.append(area_doublePeak_fit, float(row['area_doublePeak_fit']))
        gof_biGauss = np.append(gof_biGauss, float(row['gof_biGauss']))
        gof_doublePeak = np.append(gof_doublePeak, float(row['gof_doublePeak']))
        
```

```python
'''
Checking the number of DOMs that fail in minimizing
'''

success_rate_biGauss = len(success_biGauss[success_biGauss == 'True'])/float(len(DOM)) * 100
success_rate_doublePeak = len(success_doublePeak[success_doublePeak == 'True'])/float(len(DOM)) * 100

success_rate = len(success_biGauss[(success_biGauss == 'True') & (success_doublePeak == 'True')])/float(len(DOM)) * 100
print(success_rate_biGauss, success_rate_doublePeak, success_rate)
```

```python
'''
Removing Failed Events
'''
min_bool = (success_biGauss == 'True') & (success_doublePeak == 'True')
file_num = file_num[min_bool]
frame_num = frame_num[min_bool]
lepton = lepton[min_bool]                  
DOM = DOM[min_bool]
string = string[min_bool]
binEntries_mean = binEntries_mean[min_bool]
biGauss_pos = biGauss_pos[min_bool]
biGauss_wid = biGauss_wid[min_bool]
biGauss_rat = biGauss_rat[min_bool]
biGauss_amp = biGauss_amp[min_bool] 
doublePeak_pos1 = doublePeak_pos1[min_bool]
doublePeak_wid1 = doublePeak_wid1[min_bool]
doublePeak_rat1 = doublePeak_rat1[min_bool]
doublePeak_amp1 = doublePeak_amp1[min_bool]
doublePeak_pos2 = doublePeak_pos2[min_bool] 
doublePeak_wid2 = doublePeak_wid2[min_bool]
doublePeak_rat2 = doublePeak_rat2[min_bool] 
doublePeak_amp2 = doublePeak_amp2[min_bool] 
area_data = area_data[min_bool]
area_biGauss_fit = area_biGauss_fit[min_bool] 
area_doublePeak_fit = area_doublePeak_fit[min_bool] 
gof_biGauss = gof_biGauss[min_bool]
gof_doublePeak = gof_doublePeak[min_bool]
```

```python
'''
Area under the curve - ratio
'''

area_data_biGauss = area_biGauss_fit/area_data
area_data_dp = area_doublePeak_fit/area_data

plt.figure(figsize=(10,9))
bins2 = np.arange(min(area_data_dp), max(area_data_dp), 0.1)
plt.hist(area_data_dp, bins=bins2, histtype='step', linewidth=2, color='r', log=True, label = 'Double Peak')
bins1 = np.arange(min(area_data_biGauss), max(area_data_biGauss), 0.1)
plt.hist(area_data_biGauss, bins = bins1, histtype='step', linewidth=2, log=True, label = 'Single Peak')
plt.legend(fontsize = 16)
plt.xlabel('data/fit', fontsize = 18)
plt.title('Area under the curve', fontsize = 22)
```

```python
plt.figure(figsize=(10,9))
binsX = np.arange(min(np.log10(area_data)), max(np.log10(area_data)), 0.5)
binsY = np.arange(min(area_data_dp[area_data_dp > 3]), max(area_data_dp), 0.05)
plt.hist2d(np.log10(area_data), area_data_dp, bins=[binsX, binsY], cmap=plt.cm.summer)
plt.xlabel('total Hits', fontsize = 18)
plt.ylabel('data/(double peak fit)', fontsize = 18)
plt.title('Area under the curve', fontsize=20)
plt.colorbar()
```

```python
'''
Goodness of Fit
'''

'''
BiGauss
'''
plt.figure(figsize=(10,10))
plt.hist(gof_biGauss, bins=10, log=True)
plt.xlabel('chi2/dof', fontsize=18)
plt.title('Goodness of fit - Single Peak', fontsize=20)

plt.figure(figsize=(10,10))
plt.hist(gof_biGauss[gof_biGauss < 1.5], bins=10, log=True)
plt.xlabel('chi2/dof', fontsize=18)
plt.title('Goodness of fit - Single Peak', fontsize=20)


'''
Double Peak
'''
plt.figure(figsize=(10,10))
plt.hist(gof_doublePeak[gof_doublePeak == gof_doublePeak], bins= 10, log=True)
plt.xlabel('chi2/dof', fontsize=18)
plt.title('Goodness of fit - Double Peak', fontsize=20)

plt.figure(figsize=(10,10))
plt.hist(gof_doublePeak[(gof_doublePeak == gof_doublePeak)&(gof_doublePeak<1.5)], bins= 10, log=True)
plt.xlabel('chi2/dof', fontsize=18)
plt.title('Goodness of fit - Double Peak', fontsize=20)

print(len(gof_biGauss), len(gof_biGauss[gof_biGauss < 1.5]), len(gof_doublePeak[gof_doublePeak == gof_doublePeak]), len(gof_doublePeak[(gof_doublePeak == gof_doublePeak)&(gof_doublePeak<1.5)]))
print('Common Events - ', len(gof_biGauss[(gof_biGauss < 1.5)&(gof_doublePeak == gof_doublePeak)&(gof_doublePeak<1.5)]))

'''
Plot 2D Histogram
'''

plt.figure(figsize=(10,9))
binsX = np.arange(min(area_data_dp[(gof_doublePeak == gof_doublePeak)&(gof_doublePeak<1.5)]), max(area_data_dp[(gof_doublePeak == gof_doublePeak)&(gof_doublePeak<1.5)]), 0.005)
binsY = np.arange(min(np.log10(gof_doublePeak[(gof_doublePeak == gof_doublePeak)&(gof_doublePeak<1.5)])), max(np.log10(gof_doublePeak[(gof_doublePeak == gof_doublePeak)&(gof_doublePeak<1.5)])), 0.05)
plt.hist2d(area_data_biGauss[(gof_doublePeak == gof_doublePeak)&(gof_doublePeak<1.5)], np.log10(gof_biGauss[(gof_doublePeak == gof_doublePeak)&(gof_doublePeak<1.5)]), bins=[binsX, binsY], cmap=plt.cm.summer)
plt.xlabel('area ratio - (data/double Peak fit)', fontsize = 16)
plt.ylabel('gof double peak - (chi2/dof)', fontsize = 16)
plt.title('Area under the curve', fontsize=20)
plt.colorbar()
```

```python
'''
Time Difference
'''

TimeDifference = abs(doublePeak_pos1 - doublePeak_pos2)
eMinus_DOMS = TimeDifference[(lepton == 'EMinus')&(gof_doublePeak == gof_doublePeak)&(gof_doublePeak<1.5)]
ePlus_DOMS = TimeDifference[(lepton == 'EPlus')&(gof_doublePeak == gof_doublePeak)&(gof_doublePeak<1.5)]
tMinus_DOMS = TimeDifference[(lepton == 'TauMinus')&(gof_doublePeak == gof_doublePeak)&(gof_doublePeak<1.5)]
tPlus_DOMS = TimeDifference[(lepton == 'TauPlus')&(gof_doublePeak == gof_doublePeak)&(gof_doublePeak<1.5)]

eMinus_DOMS = TimeDifference[(lepton == 'EMinus')]
ePlus_DOMS = TimeDifference[(lepton == 'EPlus')]
tMinus_DOMS = TimeDifference[(lepton == 'TauMinus')]
tPlus_DOMS = TimeDifference[(lepton == 'TauPlus')]

e = np.append(eMinus_DOMS, ePlus_DOMS)
tau = np.append(tMinus_DOMS, tPlus_DOMS)

plt.figure(figsize=(10,9))
bins = np.arange(0, 1e6, 1)
_ = plt.hist(e, bins=bins, histtype='step', log=True, label = 'e')
_ = plt.hist(tau, bins=bins, histtype='step', log=True, color='r', label = 'tau')
plt.title('TIME DIFFERENCE', fontsize = 20)
plt.legend(fontsize = 16)
plt.xlabel('Time Difference[ns]', fontsize = 18)
```
